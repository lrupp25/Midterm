## Feature Engineering

Develop any new feature(s) that you feel may be relevant to a model. Provide a narrative explanation of your choices to accompany any code.

To help, I've included a `column_builder()` utility function that will create a new boolean column based on whether a string of text appears in any of (1) the recipe title; (2) the recipe description; or (3) the recipe tags.

```python
import numpy as np
```

```python
data.dtypes
```

```python
#column_builder was never meant to be
```

```python
def column_builder(category: str, dataset: pd.DataFrame) -> pd.DataFrame:
    dataset[f'is_{category}'] = (
        #dataset['categories'].apply(
          #  lambda x: category in x
       # ),
        dataset['desc'].apply(
            lambda x: category in x
        ),
        dataset['title'].apply(
            lambda x: category in x
        )).astype(int)

    return dataset


categories = [
    'easy',
    'breakfast'
    # Add any additional keywords here
]

#data['categories'] = data['categories'].replace(np.NaN, [])
data['desc'] = data['desc'].replace(np.NaN, '')
data['title'] = data['title'].replace(np.NaN, '')


for category in categories:
    data = column_builder(category, data)

data['is_easy'].describe()
```

```python
data.head()
```

Firstly, I will make a column that will give a boolean (.astype(int)), of if any of fat, calories, protein, sodium are NaN, or non existent. I wanted to do this to check if the reviewers may leave a lower rating simply because it was missing nutrient information.

```python
data['is_missing_info'] = ((data['fat'].isnull() == True) |
                           (data['calories'].isnull() == True) |
                           (data['protein'].isnull() == True) |
                           (data['sodium'].isnull() == True)).astype(int)
data.head()
```

Next, I want to transform the directions and ingredients columns into the lengths of each object rather than the string descriptions, this is because for me I feel that length of either ingredients or directions could make someone following the instructions skip a couple steps or ingredients, thus ruining the dish as the poster intended.

```python
data.dtypes
```

the dtypes as above was only because for some reason the directions column had turned into type int64 somehow my first run through. Not sure at all how.

```python
data['dlength'] = data['directions'].str.len()
data['dlength'].describe()
```

```python
data['i_length'] = data['ingredients'].str.len()
data['i_length'].describe()
```

```python
data.head(50)
```

Imputation of Null values using the mean, can still test to see differences in null value due to prior feature engineering.

Side Note: I am surprised how much easier imputation is in pandas as compared to R.

```python
data['dlength'].fillna(data['dlength'].mean(), inplace = True)
data['i_length'].fillna(data['i_length'].mean(), inplace = True)
data['rating'].fillna(data['rating'].mean(), inplace = True)

```

Creating of bins, only added after the original one gave an R of .03

```python
data['rating_bins'] = pd.cut(x=data['rating'], bins=[0, 1, 2, 3, 4, 5],
                             labels=['0-1','1-2','2-3','3-4','4-5'])
data['rating_bins'].describe()
```

```python
print(data.isnull().sum())
```

```python
data.dropna(subset=['rating_bins'],inplace = True)
```

```python
print(data.isnull().sum())
```

## Model Building

Build a model (either a regression or a neural network) to predict a recipe's rating based on any relevant attributes that you defined in the prior steps.

You may choose to predict rating as a continuous value (0.0 to 5.0), or as a categorical (low/medium/high or similar).

Provide a narrative explanation of your choices to accompany any code.

```python
model_data = data.loc[:,['rating_bins', 'is_missing_info','dlength','i_length']]
model_data.dtypes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    model_data,
    train_size=0.8,
    random_state=42)

model = LogisticRegression(multi_class='multinomial').fit(
    X=train_data.loc[:, [
       'is_missing_info',
        'dlength','i_length']],
    y=train_data['rating_bins']
)
model
```

## Model Evaluation

After training your model, evaluate its performance. What metric(s) did you choose to optimize on? Would you say that your model performed well or poorly? How did you evaluate its performance to arrive at that conclusion?

```python
model.score(
    X=test_data.loc[:, ['is_missing_info',
        'dlength','i_length'
        ]],
    y=test_data['rating_bins']
)
```
